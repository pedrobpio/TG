```python
# importa, seta configuracao, le arquivo e tira colunas 'unnamed'
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
import seaborn as sns

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
df = pd.read_csv('dados_filtrados_v4_cotistas.csv')
```

```python
df.columns
```

```python
f = (
    df.loc[:, ['Sexo', 'Tipo de Escola', 'Raça', 'Curso', 'Créditos cursados no total', 'Créditos cursados no semestre(com aprovação)',
              'Média do período', 'Menção na disciplina', 'Código da disciplina','semestre', 'idade', 'saiu']]
        .dropna()
).corr()
plt.figure(figsize=(16, 6))
sns.heatmap(f, annot=True)
```

```python
def drawHeatMap(df):
    f = (
    df.loc[:, ['Sexo', 'Tipo de Escola', 'Raça', 'Créditos cursados no total', 'Créditos cursados no semestre(com aprovação)',
              'Média do período', 'Menção na disciplina', 'Código da disciplina','semestre', 'idade', 'saiu']]
        .dropna()
    ).corr()
    plt.figure(figsize=(16, 6))
    ax = plt.axes()
    sns.heatmap(f, annot=True, ax = ax)
    ax.set_title(df['Curso'].values[0])

def drawHeatMapCov(df):
    f = (
    df.loc[:, ['Sexo', 'Tipo de Escola', 'Raça', 'Créditos cursados no total', 'Créditos cursados no semestre(com aprovação)',
              'Média do período', 'Menção na disciplina', 'Código da disciplina','semestre', 'idade', 'saiu']]
        .dropna()
    ).cov()
    plt.figure(figsize=(16, 6))
    ax = plt.axes()
    sns.heatmap(f, annot=True, ax = ax)
    ax.set_title(df['Curso'].values[0])
```

```python
df0 = pd.read_csv('dados_filtrados_v4_cotistas_0.csv')
df1 = pd.read_csv('dados_filtrados_v4_cotistas_1.csv')
df2 = pd.read_csv('dados_filtrados_v4_cotistas_2.csv')
df3 = pd.read_csv('dados_filtrados_v4_cotistas_3.csv')
df4 = pd.read_csv('dados_filtrados_v4_cotistas_4.csv')
df5 = pd.read_csv('dados_filtrados_v4_cotistas_5.csv')
```

```python
drawHeatMap(df0)
drawHeatMap(df1)
drawHeatMap(df2)
drawHeatMap(df3)
drawHeatMap(df4)
drawHeatMap(df5)
```
