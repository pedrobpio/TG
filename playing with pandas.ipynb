```python
def converteSemeste(periodoIngresso, anoSemestreCorrente):
    anoPeriodo = int(str(periodoIngresso)[:-1])
    numPeriodo = int(str(periodoIngresso)[-1:])

    anoSemestre = int(str(anoSemestreCorrente)[:-1])
    numSemestre = int(str(anoSemestreCorrente)[-1:])

    periodo = (anoSemestre - anoPeriodo) *2
    periodo = (periodo +1 if numSemestre == 2 else periodo)
    periodo = (periodo - 1 if numPeriodo == 2 else periodo)
    periodo = (periodo - 0.5 if numSemestre == 0 else periodo)
    return periodo
```

```python
import pandas as pd
import seaborn as sns
from pandas.tools.plotting import radviz
# We'll also import seaborn, a Python graphing library
import warnings # current version of seaborn generates a bunch of warnings that we'll ignore
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA as sklearnPCA

sns.set(style="white", color_codes=True)

# Next, we'll load the Iris flower dataset, which is in the "../input/" directory
iris = pd.read_csv("iris.csv") # the iris dataset is now a Pandas DataFrame

X_std = iris.ix[:,0:4].values
y = iris.ix[:,4].values
# Let's see how many examples we have of each species
sklearn_pca = sklearnPCA(n_components=2)
Y_sklearn = sklearn_pca.fit_transform(X_std)

with plt.style.context('seaborn-whitegrid'):
    plt.figure(figsize=(6, 4))
    for lab, col in zip(('setosa', 'versicolor', 'virginica'),
                        ('blue', 'red', 'green')):
        plt.scatter(Y_sklearn[y==lab, 0],
                    Y_sklearn[y==lab, 1],
                    label=lab,
                    c=col)
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.legend(loc='lower center')
    plt.tight_layout()
    plt.show()


plt.show()
```
